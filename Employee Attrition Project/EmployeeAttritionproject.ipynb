```python
# Attrition Distribution
attrition = df['Attrition'].value_counts(normalize=True)
plt.figure(figsize=(6,5))
ax = sns.barplot(x=attrition.index, y=attrition, palette = 'deep', hue= attrition, legend=False)
for p in ax.patches:
    ax.annotate(f'{p.get_height() * 100:.2f}%',
                (p.get_x() + p.get_width() / 2.,
p.get_height()),
                ha='center', va='bottom')
plt.title('Distribution of Attrition Rate')
plt.xlabel('Attrition')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()
```

```python
# Visualizations :-
# Attrition by Education (
plt.figure(figsize=(9,5))
AttrbyEdu.plot(kind= 'pie', autopct= '%1.1f%%', startangle= 90)
centre_circle = plt.Circle((0,0) , 0.7, color = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Attrition by Education')
plt.show()
```

```python
#  Attrition by Gender (TreeMap)
plt.figure(figsize=(5,5))
labels= [f"{label}\n{value}" for label,
         value in zip(AttrbyGender.index,AttrbyGender.values)]
squarify.plot(sizes = AttrbyGender.values,label = labels ,alpha= 0.8)
plt.axis('off')
plt.title('Attrition by Gender')
plt.show()
```

```python
# Attrition by Age
plt.figure(figsize=(6,5))
sns.histplot( data = df[df['Attrition']=='Yes'], x='Age', kde= True)
plt.title('Attrition Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Attrition by JobRole
plt.figure(figsize=(6,5))
sns.countplot( data = df[df['Attrition']=='Yes'], y='JobRole', hue='JobRole')
plt.title('Attrition Distribution by JobRole')
plt.xlabel('Count')
plt.ylabel('JobRole')
plt.show()
```

```python
# Attrition by Department and Salary Hike Percent
plt.figure(figsize=(15,6))
ax1= plt.subplot(1, 2, 1)
sns.countplot(data = df[df['Attrition']=='Yes'], x= 'Department', ax=ax1, hue= 'Department')
ax1.set_title('Attrition by Department')
ax1.set_xlabel('Department')
ax1.set_ylabel('Count')
ax2= plt.subplot(1, 2, 2)
sns.histplot( data = df[df['Attrition']=='Yes'], x='PercentSalaryHike', kde= True)
ax2.set_title('Attrition Distribution by Salary Hike %')
ax2.set_xlabel('Salary Hike %')
ax2.set_ylabel('Count')
plt.tight_layout()
plt.show()
```
